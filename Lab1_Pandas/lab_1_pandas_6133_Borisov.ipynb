##### 2.9.2 через DataFrame.apply

```python
def get_seconds(row):
    time_split = row['Length'].split(':')
    return int(time_split[0]) * 60 + int(time_split[1])
    
data["Seconds"] = data.apply(get_seconds, axis=1)
```

##### 2.9.3 через Series.apply

```python
data["Minutes"] =  data["Seconds"].apply(lambda s: s / 60)
```

### 3 Продемонстрировать работу .groupby, на основе группировок в groupby вычисляете агрегатные функции по одной или нескольким колонкам. Какую информацию о данных вы получили, выполнив данную операцию? Запишите ответ в комментариях.



Время видео для каждой категории

```python
top_categories = data[['Category', 'Seconds', 'Minutes']].groupby('Category').sum().sort_values(by="Category")
top_categories
```

### 4 Решейпинг данных 1Dto2D с помощью .pivot (можно подать на вход результаты агрегатов, полученных ранее через .groupby (сгруппировать по двум полям), либо прекрасно заходит сюда данные из SQL сгруппированные предварительно по 2-м полям)


Долго думал как применить .pivot для результата .groupby выше, но не смог придумать что-то полезное. Пускай будет "вывод времени в минутах для категории только в той колонке секунд, которая соответствует категории"

```python
top_categories.pivot(columns="Seconds", values='Minutes').fillna(0)
```

### 5 Решейпинг 1Dto2D данных соединённый с группировкой / агрегацией (одним словом — сводная таблица): .pivot_table


Количество видео, созданных на канале, принадлежавших определённой категории
(Например, канал 3Blue1Brown создал 30 видео категории Science)

```python
pd.pivot_table(data, values='Id', index='Category', columns='Channel', aggfunc="count").fillna(0)
```
