```python
df = data.copy()
df.shape
```

```python
# Генерация Nan в случайные ячейки столбца CC
def gen_na(df, size=10):
    rand_index = np.random.randint(data.shape[0], size=size)
    df["CC"][rand_index] = np.nan
    
gen_na(df)
```

```python
# Удаление строк с Nan в столбце CC
df.dropna(inplace=True, subset='CC')
df.shape
```

```python
# Генерация Nan и замена их на среднее значение колонки СС
df = data.copy()
gen_na(df, size=15)
m = data['CC'].mean()
df["CC"].fillna(m, inplace=True)
df[df['CC'] == m].head(5)
```

##### 2.9.2 через DataFrame.apply

```python
def get_seconds(row):
    time_split = row['Length'].split(':')
    return int(time_split[0]) * 60 + int(time_split[1])
    
data["Seconds"] = data.apply(get_seconds, axis=1)
```

### 14 Проитерировать dataframe построчно .iterrows(). Выполните что-то имеющее смысл внутри цикла

```python
# Найдём количество и индекс каждого видео выбранной категории
category = 'Blog'
count = 0
for i, row in data.iterrows():
    if row['Category'].find(category) != -1:
        count += 1
        print(f'Найдено видео категории {category} с индексом {i}')
print(f'Количество видео: ', count)
```
